## Data Cleaning

```python
df.columns = [
    'Month', 'Bananas', 'Oranges', 'Bread', 'Tomatoes', 
    'Chicken', 'Electricity', 'Eggs', 'Gasoline', 'GroundChuck', 
    'UtilityGas', 'Milk'
]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#fill missing numeric values with the mean
df.fillna(df.mean(), inplace=True)
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Bananas', data=df)
plt.title('Time Series: Bananas Price over Time')
plt.show()
```

```python
# Plot time series for all columns
plt.figure(figsize=(15, 72))

for column in df.columns[1:]:  # Exclude 'Month' column
    plt.subplot(12, 1, df.columns.get_loc(column))
    sns.lineplot(x='Month', y=column, data=df)
    plt.title(f'Time Series: {column} over Time')
```

```python
# Plot time series for all columns
plt.figure(figsize=(15, 36))

for column in df.columns[1:]:  # Exclude 'Month' column
    plt.subplot(6, 2, df.columns.get_loc(column))
    sns.lineplot(x='Month', y=column, data=df)
    plt.title(f'Time Series: {column} over Time')
```

```python
# 4. Correlation Analysis
print("\n4. Correlation Analysis:")
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Price Comparison for Groceries
groceries_columns = ['Bananas', 'Bread', 'Tomatoes', 'Chicken', 'Eggs', 'GroundChuck', 'Milk']

plt.figure(figsize=(12, 6))
for column in groceries_columns:
    sns.lineplot(x='Month', y=column, data=df, label=column)

plt.title('Groceries Price Comparison over Time')
plt.legend()
plt.show()
```

```python
# Price Comparison for Utilities
utilities_columns = ['Electricity', 'Gasoline', 'UtilityGas']

plt.figure(figsize=(12, 6))
for column in utilities_columns:
    sns.lineplot(x='Month', y=column, data=df, label=column)

plt.title('Utilities Price Comparison over Time')
plt.legend()
plt.show()
```

```python
# 6. Distribution Plots
print("\n6. Distribution Plots:")
plt.figure(figsize=(12, 6))
sns.histplot(df['Bananas'], kde=True, bins=20)
plt.title('Distribution of Bananas Prices')
plt.show()
```

```python
# 6. Distribution Plots for All Columns
print("\n6. Distribution Plots for All Columns:")
plt.figure(figsize=(15, 10))

for column in df.columns[1:]:  # Exclude 'Month' column
    plt.subplot(3, 4, df.columns.get_loc(column))
    sns.histplot(df[column], kde=True, bins=20)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# Outlier Detection for All Columns in One Plot
print("\nOutlier Detection for All Columns in One Plot:")
plt.figure(figsize=(15, 10))

# Use a single subplot for all boxplots
sns.boxplot(data=df.drop(columns='Month'), width=0.8, palette='colorblind')

plt.title('Boxplots of Numeric Columns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 8. Comparative Analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Bananas', data=df, label='Bananas')
sns.lineplot(x='Month', y='Oranges', data=df, label='Oranges')
# Repeat for other variables...
plt.title('Comparative Analysis over Time')
plt.show()
```

## Prediction Analysis

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

time_series_column = 'Bananas'

# Split the data into training and testing sets
train_size = int(len(df) * 0.8)
train, test = df[time_series_column][:train_size], df[time_series_column][train_size:]

# Fit ARIMA model
order = (1, 1, 1)  # Adjust order based on model performance
model = ARIMA(train, order=order)
fit_model = model.fit()

# Forecast future values
future_steps = len(test)  # Adjust as needed
forecast = fit_model.forecast(steps=future_steps)

# Evaluate model performance
rmse = sqrt(mean_squared_error(test, forecast))
print(f'\nRoot Mean Squared Error (RMSE): {rmse}')

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(test.index, forecast, label='Forecast', linestyle='dashed')
plt.title(f'ARIMA Forecasting for {time_series_column}')
plt.legend()
plt.show()
```
